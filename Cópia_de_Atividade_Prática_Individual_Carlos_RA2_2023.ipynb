1) $f(x) = (x^2 - 4)(x^3 - 7x + 1)$

*Gráfico*

```python
import plotly.graph_objs as go
import numpy as np

# Definindo a função 
def f(x):
    return (x**2 - 4) * (x**3 - 7*x + 1)

# Definindo as linhas
# Gerando os valores de x
x = np.linspace(-10, 10, 100)
y = f(x)

# Criando o gráfico de linha para a função e sua derivada
fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name="$f(x) = (x^2 - 4)(x^3 - 7x + 1)$"))

# Mostrando o gráfico
fig.show()
```

*Passo a Passo*

$f(x) = (x^2 - 4)(x^3 - 7x + 1) x = 5$

$\frac{d}{dx}(x^2 - 4)(x^3 - 7x + 1)$

Regra dos produtos

$u = x^2 - 4$

$v = x^3 - 7x + 1$

$(x^2 - 4)\frac{d}{dx}(x^3 - 7x + 1) + (x^3 - 7x + 1)\frac{d}{dx}(x^2 - 4)$

$(x^2 - 4)(3x^2 -7) + (x^3 - 7x + 1)(2x)$

$2x(x^3 - 7x + 1) + (x^2 - 4)(3x^2 -7) = $

$\frac{d}{dx}5x^4 - 33x^2 +2x + 28$

Agora fazemos a derivada da função no ponto para pegar a inclinação da tangente:

$m = f'(x)$ sendo x = 5

$m = 5*5^4 - 33*5^2 + 2*5 + 28$

$m = 2238$

Agora utilizamos a função $arctan(m)$ para encontrar o ângulo

$Angulo = arctan(m)$

$Angulo = 1.570349499$ ou em um ângulo de $89.97$ graus

$\frac{d}{dx}(sin(x^2)) * cos(2x) + sin(x^2) * \frac{d}{dx}(cos(2x)) = $

$\frac{d}{dx}2x * cos(x^2)cos(2x) - 2sin(2x)sin(x^2)$

Agora fazemos a derivada da função no ponto para pegar a inclinação da tangente:

$m = f'(x)$ sendo x = 5

$m = 2*5cos(5^2)cos(5*2) - 2sin(5*x)sin(5^2)$

$m = 8.77$

$Angulo = arctan(m)$

$Angulo = 1.46$ ou em um ângulo de $84$ graus

# Questão 3 (vale 2 pontos)
Considerando a função $f(x,y) = \sin(xy) + \cos(x+y)$, no intervalo $0\leq x\leq 5 \space\space e \space\space 0\leq y\leq 5$ trace, usando o plotly o gráfico desta função e marque os pontos de máximo e mínimo neste intevalo.

Lembre-se o sympy é sempre uma opção e você pode usar qualquer técnica, algoritmo, regra, ou tabela, que tenha sido usada na matéria. E que precisa de apenas uma casa decimal de precisão. Uma dica importante para a solução deste problema é a função **np.unravel_index**


*Gráfico*

```python
import numpy as np
import plotly.graph_objects as go

# Função de duas variáveis
def f(x, y):
    return np.sin(x*y) * np.cos(x+y)

x = np.linspace(0, 5, 200)
y = np.linspace(0, 5, 200)

# Criando a base do gráfico 3D cm sua Meshgrid
X, Y = np.meshgrid(x, y)

# O valor Z para X e Y
Z = f(X, Y)

# Encontrando os pontos de mínimo e máximo
Xmin, Ymin = np.unravel_index(np.argmin(Z), Z.shape)
Xmax, Ymax = np.unravel_index(np.argmax(Z), Z.shape)

# Criando o gráfico
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z)])

# Adicionando marcadores para os pontos de mínimo e máximo
fig.add_trace(go.Scatter3d(x=[x[Xmin]], y=[y[Ymin]], z=[Z[Xmin, Ymin]], mode='markers', marker=dict(size=10, color='red', symbol='x'), name='Mínimo'))
fig.add_trace(go.Scatter3d(x=[x[Xmax]], y=[y[Ymax]], z=[Z[Xmax, Ymax]], mode='markers', marker=dict(size=10, color='green', symbol='x'), name='Máximo'))

# Detalhes finais
fig.update_layout(title='Gráfico da função f(x,y) = sin(xy) + cos(x+y)', scene=dict(xaxis_title='x', yaxis_title='y', zaxis_title='f(x,y)'))

# Mostrando o gráfico
fig.show()
```
